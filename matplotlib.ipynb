```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x = [1,2,3,4,5,6]
y= [50,45,48,47,49,45]
plt.plot(x,y, color='blue', marker='D', linewidth=2 )
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.title("Weather")
```

```python
plt.plot(x,y, 'rD', linewidth=2 )
```

```python
plt.plot(x,y, color='blue', marker='D', linewidth=2 , markersize = 14)
```

```python
plt.plot(x,y, color='blue', marker='D', linewidth=2, alpha = .7, markersize= 15)
```

```python
days=[1,2,3,4,5,6,7]
max_n = [50,51,52,48,47,49,53]
min_n = [43,44,45,46,42,41,40]
avg_n = [45,48,47,44,44,45,42]
plt.plot(days,max_n, label='max')
plt.plot(days,min_n, label='min')
plt.plot(days,avg_n, label='avg')

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.legend(loc='best', shadow=True, fontsize='small')
plt.grid()
```
