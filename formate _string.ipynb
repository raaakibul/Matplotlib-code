```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
temp = [45,51,47,52,48,51,46]
```

```python
plt.title('Weather')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(days,temp, 'green')
```

```python
plt.plot(days,temp, 'g+')
```

```python
plt.plot(days,temp, 'g+--')
```

```python
plt.plot(days,temp, 'g+-.')
```

```python
plt.plot(days,temp, 'r*')
```

```python
plt.plot(days,temp, color='red', marker='D', linestyle='', markersize=4)
```

```python
plt.plot(days,temp, color='red', marker='D', linestyle='', markersize=4, alpha=.7)
```
