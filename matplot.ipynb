```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = [1,2,3,4,5,6,7]
y= [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,color= 'red', linewidth=3,linestyle= 'dotted')
plt.ylabel("Temprature")
plt.xlabel("Day")
plt.title("Weather")
```

```python

plt.ylabel("Temprature")
plt.xlabel("Day")
plt.title("Weather")
plt.plot(x,y, 'r.')
```

```python

days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [45,43,42,46,47,41,40]
avg_t = [46,44,45,47,41,40,45]

```

```python
plt.xlabel("Days")
plt.ylabel("Temparture")
plt.title("Weather")
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend()
```
