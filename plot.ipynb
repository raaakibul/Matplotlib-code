```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [50,45,52,49,52,47,51]
 
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(x,y, 'rD', color= "green" )
```

```python
plt.plot(x,y, color="red", marker='D', linestyle='')
```

```python
plt.plot(x,y,color= 'blue')
```

```python
plt.plot(x,y,color='green', marker='+', linestyle='', markersize=20)
```

```python
plt.plot(x,y, color="green", alpha=.5)
```
